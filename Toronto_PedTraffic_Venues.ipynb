# Define function to extract distance of a venue to a crossing

```python
def get_dist(dframe_ven,s_lat,s_lon):
    nearby_ven = dframe_ven
    nearby_ven['latdiff'] = nearby_ven['venue.location.lat'] - s_lat
    nearby_ven['londiff'] = nearby_ven['venue.location.lng'] - s_lon
    nearby_ven['latdiff'] = nearby_ven['latdiff']**2
    nearby_ven['londiff'] = nearby_ven['londiff']**2
    nearby_ven['totaldiff'] = nearby_ven['latdiff'] + nearby_ven['londiff']
    nearby_ven['totaldiff'] = nearby_ven['totaldiff']**(1/2)
    nearby_ven.sort_values('totaldiff', ascending=False, inplace=True)
    return nearby_ven
```

### Normalize y

```python
import math
def roundup(x):
    return int(math.ceil(x / 10000.0)) * 10000
```

```python
def categorize_y(y_arr):
    buckets = 5
    maxim = roundup(max(y_arr))
    minim = 0
    bu1 = maxim-(maxim-minim)/5*1
    bu2 = maxim-(maxim-minim)/5*2
    bu3 = maxim-(maxim-minim)/5*3
    bu4 = maxim-(maxim-minim)/5*4
    bu5 = maxim-(maxim-minim)/5*5
    out_arr = np.empty(len(y_arr), dtype=int)
    for ind in range(len(y_arr)):
        #print(ind,"   ",y_arr[ind])
        if y_arr[ind] > bu1:
            out_arr[ind] = "5"
        elif y_arr[ind] > bu2:
            out_arr[ind] = "4"
        elif y_arr[ind] > bu3:
            out_arr[ind] = "3"
        elif y_arr[ind] > bu4:
            out_arr[ind] = "2"
        elif y_arr[ind] >= 0:
            out_arr[ind] = "1"
        print(out_arr[ind],y_arr[ind])
    return out_arr           
        
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
#split into train and test set
X_train, X_testset, y_train, y_testset = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_testset.shape,  y_testset.shape)
```

```python
y_testset
```

```python

#find the best k
Ks = 50
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_testset)
    mean_acc[n-1] = metrics.accuracy_score(y_testset, yhat)

    
    std_acc[n-1]=np.std(yhat==y_testset)/np.sqrt(yhat.shape[0])

mean_acc
print( "The best accuracy is", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```

```python
#retrain with best k
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = 2).fit(X_train,y_train)
neigh
```
